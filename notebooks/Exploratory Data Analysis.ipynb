```python
# Specialization

'''Converting all the different specializations into 8 broad types as many of these 
specializations are quite similiar'''

cs_subs = ['computer engineering','computer application','computer science and technology','computer science','computer networking','computer and communication engineering','software engineering']
ece_subs = ['electronics and telecommunication','electronics and instrumentation eng','electronics and instrumentation engineering','electronics engineering','applied electronics and instrumentation','telecommunication engineering','electronics and computer engineering','computer and communication engineering','communication engineering','electronics','electronics & instrumentation eng','electronics & telecommunications','instrumentation and control engineering','information & communication technology']
eee_subs = ['electrical engineering','electrical and power engineering','embedded systems technology']
me_subs =  ['industrial & production engineering','mechanical and automation','automobile/automotive engineering','instrumentation engineering','mechatronics','aeronautical engineering','industrial engineering','mechanical & production engineering','industrial & management engineering','control and instrumentation engineering','vlsi design and cad','operational research','aerospace engineering','cad / cam','industrial engineering and management','internal combustion engine','power systems and automation']
bt_subs = ['biotechnology','biomedical engineering' ,'environment science']
meta_subs = ['metallurgical engineering' , 'ceramic engineering' , 'polymer technology' , 'textile engineering']
it_subs = ['information science','information science engineering']
civil_subs = ['construction technology and management']

def change_spec(data):
    data=data.apply(lambda x:'computer science & engineering' if x in cs_subs else x)
    data=data.apply(lambda x:'electronics and communication engineering' if x in ece_subs else x)
    data=data.apply(lambda x:'electronics and electrical engineering' if x in eee_subs else x)
    data=data.apply(lambda x:'metallurgical engineering' if x in meta_subs else x)
    data=data.apply(lambda x:'mechanical engineering' if x in me_subs else x)
    data=data.apply(lambda x:'biotechnology' if x in bt_subs else x)
    data=data.apply(lambda x:'information technology' if x in it_subs else x)
    data1=data.apply(lambda x:'civil engineering' if x in civil_subs else x)
    return data1

fig,ax = plt.subplots(figsize =(15,8))
ax.set(xlabel = 'Specialization',ylabel= 'Salary',title = 'Salaries of Candidates based on Specialization')
plt.xticks(rotation=90,fontsize=10)
sns.boxplot(x = change_spec(train['Specialization']),y= train['Salary'])
```

AMCAT Personality Tests

These are 5 personality traits which can be determined by the specific tests.

For a clarity on what the traits mean on which the scores are reported, see https://en.wikipedia.org/wiki/Big_Five_personality_traits.

The five traits are:

    openness to experience (inventive/curious vs. consistent/cautious)
    conscientiousness (efficient/organized vs. extravagant/careless)
    extraversion (outgoing/energetic vs. solitary/reserved)
    agreeableness (friendly/compassionate vs. challenging/callous)
    neuroticism (sensitive/nervous vs. resilient/confident)

The scores are sampled from a distribution with mean 0 and standard deviation 1


```python
fig,ax = plt.subplots(2,3,figsize =(15,10))
sns.regplot(x =train['conscientiousness'] ,y= train['Salary'] ,line_kws={"color":'red'},ax=ax[0,0])
ax[0,0].set(title = 'conscientiousness')
sns.regplot(x =train['agreeableness'] ,y= train['Salary'] ,line_kws={"color":'red'},ax=ax[0,1])
ax[0,1].set(title = 'agreeableness')
sns.regplot(x =train['extraversion'] ,y= train['Salary'] ,line_kws={"color":'red'},ax=ax[0,2])
ax[0,2].set(title = 'extraversion')
sns.regplot(x =train['nueroticism'] ,y= train['Salary'] ,line_kws={"color":'red'},ax=ax[1,0])
ax[1,0].set(title = 'nueroticism')
sns.regplot(x =train['openess_to_experience'] ,y= train['Salary'] ,line_kws={"color":'red'},ax=ax[1,1])
ax[1,1].set(title = 'openess_to_experience')
fig.delaxes(ax[1,2]) 
```

```python
#Domain Scores

#Based on the specific scores in the Specialization tests , we'll give a specific score to 
#each of the candidates.

# Also we will define a 'number' feature which will tell us how many domain tests the candidate
# has given
train['Specialization1'] = change_spec(train['Specialization'])
test['Specialization1'] = change_spec(test['Specialization'])
b = np.array(train['Specialization1'].value_counts().index)
c = ['ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
train['number']=0
for i in range(train.shape[0]):
    a = list(train.iloc[i,20:27]).count(-1)
    train.loc[i,'number']=7-a
    
test['number']=0
for i in range(test.shape[0]):
    a = list(test.iloc[i,20:27]).count(-1)
    test.loc[i,'number']=7-a    

def change_score(train):
    
    train['score']=0
    for i in range(train.shape[0]):
        if train['number'][i]==1:
            train.loc[i,'score'] = max(list(train.iloc[i,20:27]))
        elif train['number'][i]==0:
            train.loc[i,'score']==0
        else:
            if train['Specialization1'][i]=='computer science & engineering':
                train.loc[i,'score'] = train.iloc[i,22]
            elif train['Specialization1'][i]=='electronics and communication engineering':
                train.loc[i,'score'] = max(list(train.iloc[i,[21,24,25]]))   
            elif train['Specialization1'][i]=='information technology':
                train.loc[i,'score'] = train.iloc[i,20]
            elif train['Specialization1'][i]=='electronics and electrical engineering':
                train.loc[i,'score'] = max(list(train.iloc[i,[21,24,25]]))
            elif train['Specialization1'][i]=='mechanical engineering':
                train.loc[i,'score'] = train.iloc[i,23]
            elif train['Specialization1'][i]=='civil engineering':
                train.loc[i,'score'] = train.iloc[i,26]
            elif train['Specialization1'][i]=='biotechnology':
                train.loc[i,'score'] = train.iloc[i,20]
            elif train['Specialization1'][i]=='other':
                train.loc[i,'score'] = train.iloc[i,20]
            elif train['Specialization1'][i]=='chemical engineering':
                train.loc[i,'score'] = train.iloc[i,20]
            elif train['Specialization1'][i]=='metallurgical engineering':
                train.loc[i,'score'] = train.iloc[i,20]   
    
    train.loc[train['score']==-1,'score']= train.loc[train['score']==-1].iloc[:,20:27].max(axis=1)
    train['score'] = pd.to_numeric(train['score'])
    return train

train = change_score(train)
test = change_score(test)

```

## 4. Feature Engineering<a class='anchor' id='4'></a>

```python
train = train.drop(['Specialization','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg'],axis=1)
test = test.drop(['Specialization','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg'],axis=1)
```

```python
x = train.drop(['Salary'],axis=1)
y = train['Salary']
#Defining a new feature telling us about the number of years spent in college
x['CollegeYrs'] = train['GraduationYear']-train['12graduation']
x = x.drop(['12graduation'],axis=1)
test['CollegeYrs'] = test['GraduationYear']-test['12graduation']
test = test.drop(['12graduation'],axis=1)

# Defining another feature descriing the age of candidate at graduation
year = train['DOB'].apply(lambda x:x.year)
x['AgeAtGrad'] =  train['GraduationYear']- year
x = x.drop(['DOB'],axis=1)
year = test['DOB'].apply(lambda x:x.year)
test['AgeAtGrad'] =  test['GraduationYear']- year
test = test.drop(['DOB'],axis=1)

# We will now use Column Transformer to Normalize some columns with numerical data
# and Label encode some columns with categorical features

cols_label = ['Gender','10board','12board','CollegeID','CollegeState','GraduationYear','Specialization1','AgeAtGrad','CollegeYrs','Degree']
cols_normalize = ['10percentage','12percentage','collegeGPA','English','Logical','Quant','score']
ct = ColumnTransformer(transformers=[("categorical_features", OrdinalEncoder(),cols_label),('num_features',StandardScaler(),cols_normalize)],
                      remainder="passthrough")
ct.fit(x.append(test))
x1 = pd.DataFrame(ct.transform(x))
test = pd.DataFrame(ct.transform(test))

a = ['Gender','10board','12board','CollegeID','CollegeState','GraduationYear','Specialization1','AgeAtGrad','CollegeYrs','Degree','10percentage','12percentage','collegeGPA','English','Logical','Quant','score','CollegeTier','CollegeCityTier','Domain','conscientiousness', 'agreeableness', 'extraversion',
       'nueroticism', 'openess_to_experience','number']
x1.columns=a
test.columns=a
```

## 6. Observations and Hypotheses from EDA<a class='anchor' id='6'></a>

Some observations which can be made from the above EDA:

* Male candidates generally get placements of higher salary as compared to Female candidates

* The increase in English,Logical and Quant scores of the candidates increased the salary too.
